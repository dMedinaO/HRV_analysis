- Auxiliar functions

```python
def correct_format_rmssd(value):
    value = str(value)
    value = value.strip()
    value = value.replace(" ", "")
    value = value.replace(",", ".")
    return value
```

```python
def categorize_age_group(age):
    """
    Categorizes age into defined elderly age groups based on Chilean health standards.

    Parameters:
    -----------
    age : int or float
        Age of the individual in years.

    Returns:
    --------
    str
        Age category as a string label.
    """
    if age < 60:
        return "Not elderly"
    elif 60 <= age <= 69:
        return "Young older adult"
    elif 70 <= age <= 79:
        return "Intermediate older adult"
    elif 80 <= age <= 89:
        return "Older adult"
    elif age >= 90:
        return "Very old adult"
    else:
        return "Unknown"
```

```python
def categorize_bmi_elderly(bmi):
    """
    Categorizes BMI according to Chilean recommendations for elderly adults.

    Parameters:
    -----------
    bmi : float
        Body Mass Index (BMI) value in kg/m².

    Returns:
    --------
    str
        BMI category as a string label.
    """
    if pd.isnull(bmi):
        return "Null"
    elif bmi < 23:
        return "Underweight"
    elif 23 <= bmi <= 27.9:
        return "Normal weight"
    elif 28 <= bmi <= 31.9:
        return "Overweight"
    elif bmi >= 32:
        return "Obese"
    else:
        return "Unknown"
```

```python
def categorize_map_elderly(map_value):
    """
    Categorizes Mean Arterial Pressure (MAP) in elderly individuals.

    Parameters:
    -----------
    map_value : float
        Mean Arterial Pressure (MAP) in mmHg.

    Returns:
    --------
    str
        MAP category label.
    """
    if pd.isnull(map_value):
        return "Unknown"
    elif map_value < 70:
        return "Low perfusion pressure"
    elif 70 <= map_value <= 99:
        return "Normal MAP"
    elif map_value >= 100:
        return "High MAP"
    else:
        return "Unknown"
```

- Checking null values per column

```python
data_null = []

for column in df_data.columns:
    is_null_columns=df_data[column].isna()
    counts = is_null_columns.value_counts()

    summary = {"Feature": column, "Null": 0, "Not_Null": 0}

    if False in counts.index:
        summary["Not_Null"] = counts[False]
    if True in counts.index:
        summary["Null"] = counts[True]
    
    data_null.append(summary)

df_summary_null = pd.DataFrame(data_null)
df_summary_null
```

- There are three attributes with null values, we will complete it with the average value. However, the average will be estimated based on the clusters associated with the generated categories.

- Making categories for:
    - Age
    - Pam
    - IMC

- We have not selected Bp as input to generate categories, because it is complicated to handle the variables at the same time with stricted rules. This implies that it is possible the generation of a high number of Unclassified examples.

```python
df_data["age_category"] = df_data["age"].apply(categorize_age_group)
df_data["age_category"].value_counts()
```

```python
df_data["bmi_category"] = df_data["updated_imc"].round(1).apply(categorize_bmi_elderly)
df_data["bmi_category"].value_counts()
```

```python
df_data["pam_category"] = df_data["bp_pam"].round(0).apply(categorize_map_elderly)
df_data["pam_category"].value_counts()
```

- Now, we can complete the null values 

```python
df_summary_null
```

```python
for features in ["t2m_pre_sdnn", "t2m_pre_rmssd", "t2m_pre_mean_hr"]:
    df_data[f"is_null_{features}"] = df_data[features].isna()

    filter_value = df_data[df_data[f"is_null_{features}"]]

    for index in filter_value.index:
        age_category = filter_value["age_category"][index]
        bmi_category = filter_value["bmi_category"][index]
        pam_category = filter_value["pam_category"][index]
        sex = filter_value["sex"][index]

        df_with_categories = df_data[(df_data['age_category'] == age_category) &
                                    (df_data['bmi_category'] == bmi_category) &
                                    (df_data['pam_category'] == pam_category) &
                                    (df_data['sex'] == sex)]
        
        average = df_with_categories[features].mean()

        df_data[features][index] = average
```
